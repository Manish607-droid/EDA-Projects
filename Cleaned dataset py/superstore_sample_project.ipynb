```python
# Creating a new column cost per unit

dataset['Cost Per Unit'] = (dataset['Sales']- dataset['Profit'])/dataset['Quantity']
dataset
```

```python
# Since the order date is in object so we convert it to date time

dataset['Date'] = pd.to_datetime(dataset['Order Date'], dayfirst= False)
dataset[' Order Month'] = dataset['Date'].dt.month_name()
dataset[' Order Day'] = dataset['Date'].dt.day_name()
dataset['Order Year'] = dataset['Date'].dt.year
```

```python
# Since the dates have been separated now we drop the order date, ship date and Date and other unimportant columns
dataset.drop('Order Date', axis = 1, inplace = True)
dataset.drop('Ship Date', axis = 1, inplace = True)
dataset.drop('Date', axis = 1, inplace = True)
dataset.drop('Country', axis =1 , inplace = True)
dataset.drop('Postal Code', axis = 1, inplace= True)
```

```python
dataset
```

```python
# Replacing the Same Day to Premium 
dataset['Ship Mode'] = dataset['Ship Mode'].replace('Same Day', 'Premium First Class')
# Business Justification: 'Same Day' renamed to 'Premium First Class'
# because it represents a faster tier than First Class delivery
```

```python
# Renaming the columns name
dataset = dataset.rename(columns=  {' Order Month': 'Order Month', ' Order Day': 'Order Day'})
```

```python
# Answering some business questions through the dataset
```

```python
# 1) Which region has the highest sales and profit
```

```python
highest_sales = dataset.groupby('Region')['Sales'].sum().sort_values(ascending= False)
(highest_sales)
```

```python
highest_profit = dataset.groupby('Region')['Profit'].sum().sort_values(ascending= False)
print(highest_profit)

# West Region had the highest sales and profit followed by East in second.
```

```python
# 2) Most and Least profitable category?

profitable_category = dataset.groupby('Category')['Profit'].sum().sort_values(ascending= False)
print(profitable_category)

# Technology was the most profitable category followed by Office Supplies and Furniture.
```

```python
# 3) Which customer segment drives the most revenue?

most_revenue_driver = dataset.groupby('Segment')['Sales'].sum().sort_values(ascending= False)
print(most_revenue_driver)

# Consumer sement drives the most revenue followed by Corporate and Home Office
```

```python
# 4) Which Year has peak sales?

Peak_year = dataset.groupby('Order Year')['Sales'].sum()
print(f'The peak Year of sales are: {Peak_year}')

# 2017 had the peak year followed by 2016 and 2014.
```

```python
# 5) which shipping mode is most used?

modes_used = dataset['Ship Mode'].value_counts().sort_values(ascending= False)
print(modes_used)

# Standard Class is the most used followed by Second and First Class
```

```python
# Highest sales

plt.Figure(figsize=(10,5))
sales = plt.bar(highest_sales.index, highest_sales.values, color = "#EE8F12D1")
plt.xlabel('Regions')
plt.ylabel('Sales')
plt.title('Highest Sales Per Region')
plt.show()
```

```python
# profitable category

plt.Figure(figsize=(10,5))
most_profitable_category = plt.bar(profitable_category.index,profitable_category.values, color = "#9fb419")
plt.xlabel('Categories')
plt.ylabel('Profit')
plt.title('Profitable Categories')
plt.show()
```

```python
# Most revenue driver per segment

plt.Figure(figsize=(8,8))
most_revenue_driver_segment = plt.pie(most_revenue_driver, labels = most_revenue_driver.index, autopct='%1.1f%%',startangle=90,colors = ['goldenrod','cornsilk','tan'])
plt.title('Revenue Driver Per Segment')
plt.show()
```
